```python
import zipfile
import os

zip_file_path = '/content/drive/MyDrive/CSE425 Project/data/maestro-v3.0.0-midi (2).zip'
extraction_path = '/content/drive/MyDrive/CSE425 Project/data'

os.makedirs(extraction_path, exist_ok=True)

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

print(f"'{zip_file_path}' extracted successfully to '{extraction_path}'")
```
